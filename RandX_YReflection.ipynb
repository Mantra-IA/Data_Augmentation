### **IMPORTAR LIBRERIAS NECESARIAS**

```python
import tensorflow as tf
import os
from PIL import Image
import numpy as np
```

### **PROCESO DE AUMENTO DE DATOS POR MEDIO DE RandXReflection**

```python
# Definir las rutas de las carpetas
source_folder = '/content/drive/MyDrive/Data_Augmentation/0_Data_Gen_py/RESULTS/All_Mels_2/VM'
destination_folder = '/content/drive/MyDrive/Data_Augmentation/1_Data_Aug_py/RESULTS/RandXReflectionAUG/VM'

# Crear la carpeta de destino si no existe
os.makedirs(destination_folder, exist_ok=True)

# Obtener la lista de archivos en la carpeta de origen
filenames = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]

for filename in filenames:
    # Leer la imagen
    image_path = os.path.join(source_folder, filename)
    image = tf.io.read_file(image_path)
    image = tf.image.decode_image(image, channels=3)

    # Aplicar el flip horizontal
    flipped_image = tf.image.flip_left_right(image)

    # Convertir la imagen procesada a un formato que pueda ser guardado
    flipped_image = tf.cast(flipped_image, tf.uint8)
    flipped_image = Image.fromarray(flipped_image.numpy())

    # Guardar la imagen en formato PNG en la carpeta de destino
    #save_path = os.path.join(destination_folder, os.path.splitext(filename)[0] + '.png')
    save_path = os.path.join(destination_folder, 'AUG_' + os.path.splitext(filename)[0] + '.png')
    flipped_image.save(save_path, format='PNG')

print("Proceso completado. Las im치genes modificadas se guardaron en:", destination_folder)
```

### **PROCESO DE AUMENTO DE DATOS POR MEDIO DE RandYReflection**

```python
destination_folder = '/content/drive/MyDrive/Data_Augmentation/1_Data_Aug_py/RESULTS/RandYReflectionAUG/VM'
# Crear la carpeta de destino si no existe
os.makedirs(destination_folder, exist_ok=True)

# Obtener la lista de archivos en la carpeta de origen
filenames = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]

for filename in filenames:
    # Leer la imagen
    image_path = os.path.join(source_folder, filename)
    image = tf.io.read_file(image_path)
    image = tf.image.decode_image(image, channels=3)

    # Aplicar el flip vertical
    flipped_image = tf.image.flip_up_down(image)

    # Convertir la imagen procesada a un formato que pueda ser guardado
    flipped_image = tf.cast(flipped_image, tf.uint8)
    flipped_image = Image.fromarray(flipped_image.numpy())

    # Guardar la imagen en formato PNG en la carpeta de destino
    #save_path = os.path.join(destination_folder, os.path.splitext(filename)[0] + '.png')
    save_path = os.path.join(destination_folder, 'AUG_' + os.path.splitext(filename)[0] + '.png')
    flipped_image.save(save_path, format='PNG')

print("Proceso completado. Las im치genes modificadas se guardaron en:", destination_folder)
```
